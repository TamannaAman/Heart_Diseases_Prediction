```python
import pandas as pd
import numpy as np

# Sample dataset (Replace this with your actual dataset)
Data = {
    'age': [25, 40, np.nan, 50, 60],
    'sex': [1, 0, 1, 1, 0],
    'cp': [2, 3, 1, 2, 0],
    'trestbps': [120, 130, 140, 135, 128],
    'chol': [200, np.nan, 230, 210, np.nan],
    'fbs': [0, 1, 0, 1, 0],
    'restecg': [1, 0, 1, 0, 1],
    'thalach': [150, np.nan, 160, 170, np.nan],
    'exang': [0, 1, 0, 1, 0],
    'oldpeak': [1.2, 0.5, np.nan, 2.1, np.nan],
    'slope': [2, 1, 2, 1, 0],
    'ca': [0, 1, 2, 1, 0],
    'thal': [3, 2, 3, 3, 1],
    'target': [1, 0, 1, 0, 1]
}

# Create DataFrame
# Changed 'data' to 'Data' to reference the correct variable
df = pd.DataFrame(Data)

# Identify missing values
print("Missing values before handling:\n", df.isnull().sum())

# Handling missing values
df['age'].fillna(df['age'].mean(), inplace=True)  # Replace missing age with mean
df['chol'].fillna(df['chol'].mean(), inplace=True)  # Replace missing cholesterol with mean
df['thalach'].fillna(df['thalach'].mean(), inplace=True)  # Replace missing thalach with mean
df['oldpeak'].fillna(df['oldpeak'].median(), inplace=True)  # Replace missing oldpeak with median

# Check if missing values are handled
print("\nMissing values after handling:\n", df.isnull().sum())

# Print cleaned dataset
print("\nCleaned Dataset:\n", df)
```
